```python
from pypdf import PdfReader, PdfWriter

reader = PdfReader("Modulhandbuch.pdf")
writer = PdfWriter()

# Seitenzahlen sind 1-basiert übergeben, wir brauchen 0-basiert
seiten_entfernen = [num - 1 for num in [1,2,3,4,5,22,31,40,49,52]]

for i, seite in enumerate(reader.pages):
    if i not in seiten_entfernen: 
        writer.add_page(seite)

with open("CutFile.pdf", "wb") as f:
    writer.write(f)
```
